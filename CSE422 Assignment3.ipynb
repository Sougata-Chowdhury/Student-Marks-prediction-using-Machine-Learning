```python
import random

class TreeNode:
  def __init__(self, value=None, depth=0):
    self.value = value
    self.left = None
    self.right = None
    self.depth = depth
def Tree_generation(current_depth=0, max_depth=5):
  if current_depth >= max_depth:
    value = random.choice([-1, 1])
    return TreeNode(value, current_depth)
  node = TreeNode(None, current_depth)
  node.left = Tree_generation(current_depth + 1, max_depth)
  node.right = Tree_generation(current_depth + 1, max_depth)
  return node
def alpha_beta_pruning(node, alpha, beta, maximizing_player):
  if node.left is None and node.right is None:
    return node.value
  if maximizing_player:
    max_eval = float('-inf')
    if node.left:
      eval_left = alpha_beta_pruning(node.left, alpha, beta, False)
      max_eval = max(max_eval, eval_left)
      alpha = max(alpha, eval_left)
      if beta <= alpha:
        return max_eval
    if node.right:
      eval_right = alpha_beta_pruning(node.right, alpha, beta, False)
      max_eval = max(max_eval, eval_right)
      alpha = max(alpha, eval_right)
      if beta <= alpha:
        return max_eval
    node.value = max_eval
    return max_eval
  else:
    min_eval = float('inf')
    if node.left:
      eval_left = alpha_beta_pruning(node.left, alpha, beta, True)
      min_eval = min(min_eval, eval_left)
      beta = min(beta, eval_left)
      if beta <= alpha:
        return min_eval
    if node.right:
      eval_right = alpha_beta_pruning(node.right, alpha, beta, True)
      min_eval = min(min_eval, eval_right)
      beta = min(beta, eval_right)
      if beta <= alpha:
        return min_eval
    node.value = min_eval
    return min_eval
def collect_round_winner_results(node, round_winners_results, depth=1):
  if depth > 5:
    return
  if node.left is None and node.right is None:
    if node.value == -1:
      winner = "Scorpion"
    else:
      winner = "Sub-Zero"
    round_winners_results[depth].append(winner)
    return

  if node.left:
    collect_round_winner_results(node.left, round_winners_results, depth + 1)
  if node.right:
    collect_round_winner_results(node.right, round_winners_results, depth + 1)

  if node.value is not None:
    if node.value == -1:
      winner = "Scorpion"
    else:
      winner = "Sub-Zero"
    round_winners_results[depth].append(winner)

def MK_Gameplay(starting_player, total_rounds=3):
  round_winners = []
  for i in range(total_rounds):
    tree = Tree_generation(max_depth=5)
    final_result = alpha_beta_pruning(tree, float('-inf'), float('inf'), starting_player == 1)
    if final_result == -1:
      round_winners.append("Scorpion")
    else:
      round_winners.append("Sub-Zero")
    starting_player = 1 - starting_player
  print("-----------------------")
  if round_winners.count("Scorpion") > round_winners.count("Sub-Zero"):
    print("Game Winner: Scorpion")
  else:
    print("Game Winner: Sub-Zero")
  print(f"Total Rounds Played: {total_rounds}")
  round_number = 1
  for winner in round_winners:
    print(f"Winner of Round {round_number}: {winner}")
    round_number += 1
starting_player = int(input("Enter the starting player (0 for Scorpion, 1 for Sub-Zero): "))
MK_Gameplay(starting_player)
```

```python
import math

class Node:
  def __init__(self, value=None):
    self.value = value
    self.left = None
    self.right = None

def create_tree(values, max_depth, current_depth=0):
  if current_depth == max_depth:
    return Node(values.pop(0))
  node = Node()
  node.left = create_tree(values, max_depth, current_depth + 1)
  node.right = create_tree(values, max_depth, current_depth + 1)
  return node

def minimax_with_pruning(node, depth, alpha, beta, is_maximizing):
  if depth == 0 or (node.left is None and node.right is None):
    return node.value

  if is_maximizing:
    best_score = float('-inf')
    if node.left:
      left_score = minimax_with_pruning(node.left, depth - 1, alpha, beta, False)
      best_score = max(best_score, left_score)
      alpha = max(alpha, left_score)
      if beta <= alpha:
        return best_score
    if node.right:
      right_score = minimax_with_pruning(node.right, depth - 1, alpha, beta, False)
      best_score = max(best_score, right_score)
      alpha = max(alpha, right_score)
      if beta <= alpha:
        return best_score
    return best_score
  else:
    best_score = float('inf')
    if node.left:
      left_score = minimax_with_pruning(node.left, depth - 1, alpha, beta, True)
      best_score = min(best_score, left_score)
      beta = min(beta, left_score)
      if beta <= alpha:
        return best_score
    if node.right:
      right_score = minimax_with_pruning(node.right, depth - 1, alpha, beta, True)
      best_score = min(best_score, right_score)
      beta = min(beta, right_score)
      if beta <= alpha:
        return best_score
    return best_score

def maximize_only(node, depth, alpha, beta):
  if depth == 0 or (node.left is None and node.right is None):
    return node.value
  best_score = float('-inf')
  if node.left:
    left_score = maximize_only(node.left, depth - 1, alpha, beta)
    best_score = max(best_score, left_score)
    alpha = max(alpha, left_score)
    if beta <= alpha:
      return best_score
  if node.right:
    right_score = maximize_only(node.right, depth - 1, alpha, beta)
    best_score = max(best_score, right_score)
    alpha = max(alpha, right_score)
    if beta <= alpha:
      return best_score
  return best_score

def pacman_game(magic_cost):
  values = [3, 6, 2, 3, 7, 1, 2, 0]
  tree = create_tree(values, max_depth=3)
  score_without_magic = minimax_with_pruning(tree, 3, float('-inf'), float('inf'), True)
  left_subtree_score = maximize_only(tree.left, 2, float('-inf'), float('inf')) - magic_cost
  right_subtree_score = maximize_only(tree.right, 2, float('-inf'), float('inf')) - magic_cost
  score_with_magic = max(left_subtree_score, right_subtree_score)
  if right_subtree_score > left_subtree_score:
    direction = "right"
  else:
    direction = "left"
  if score_with_magic > score_without_magic:
    print(f"The new minimax value is {score_with_magic}. Pacman goes {direction} and uses dark magic.")
  else:
    print(f"The minimax value is {score_without_magic}. Pacman does not use dark magic.")
magic_cost = int(input("Enter the cost of using dark magic: "))
pacman_game(magic_cost)
```
