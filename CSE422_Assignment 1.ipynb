```python
# Task 1
def A_Star_Search(graph, heuristic_values, start, goal):
    unexplored_nodes = [start]
    fully_explored_nodes = []

    cost_from_start = {start: 0}
    parents = {start: None}

    while len(unexplored_nodes) > 0:
        current_node = None
        for node in unexplored_nodes:
            if current_node is None or cost_from_start[node] + heuristic_values[node] < cost_from_start[current_node] + heuristic_values[current_node]:
                current_node = node

        if current_node is None:
            return None

        if current_node == goal:
            path = []
            while current_node is not None:
                path.append(current_node)
                current_node = parents[current_node]
            path.reverse()
            return path, cost_from_start[goal]

        unexplored_nodes.remove(current_node)
        fully_explored_nodes.append(current_node)

        for neighbor, distance in graph[current_node].items():
            if neighbor in fully_explored_nodes:
                continue

            tentative_cost = cost_from_start[current_node] + distance

            if neighbor not in unexplored_nodes or tentative_cost < cost_from_start.get(neighbor, float('inf')):
                cost_from_start[neighbor] = tentative_cost
                parents[neighbor] = current_node

                if neighbor not in unexplored_nodes:
                    unexplored_nodes.append(neighbor)

    return None


inputfile = open('inputfile_lab1.txt', 'r')
graph = {}
heuristic_values = {}

for line in inputfile:
    line_values = line.split()
    city = line_values[0]
    heuristic = int(line_values[1])
    heuristic_values[city] = heuristic

    if city not in graph:
        graph[city] = {}

    for i in range(2, len(line_values), 2):
        neighbor = line_values[i]
        distance = int(line_values[i + 1])
        graph[city][neighbor] = distance

        if neighbor not in graph:
            graph[neighbor] = {}
        graph[neighbor][city] = distance
print(graph)
inputfile.close()

start = input("Please Enter Starting City: ")
goal = input("Please Enter Goal City: ")

result = A_Star_Search(graph, heuristic_values, start, goal)
if result!= None:
    path, cost = result
    print("Path:", " -> ".join(path))
    print("Total Cost:", cost, "km")
else:
    print('Path does not exist!')
```
